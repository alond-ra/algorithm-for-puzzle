```python
import random

def imprime(e): #imprime una lista de 16 elementos en una cuadrícula de 4x4
        print (' ')
        for i in range(0,4):
            print('|',e[i*4],'|',e[i*4+1],'|',e[i*4+2],'|',e[i*4+3],'|')
        print (' ') 
        
def h(l, final): #numeros fuera de lugar
    suma = 0
    for j in range (0,16):
        if(l[j] != final[j]):
            suma = suma + 1
    return suma

def arriba(e):
    i = e.index('__')
    if(i in [0,2,3,1]):
        print('Movimiento inválido')
        return    
    e[i] = e[i-4]
    e[i-4] = '__'
    return e

def abajo(e): 
    i = e.index('__')
    if(i in [12,13,14,15]):
        print('Movimiento inválido')
        return
    e[i] = e[i+4]
    e[i+4] = '__'
    return e


def izq(e): 
    i = e.index('__')
    if(i in [0,4,8,12]):
        print('Movimiento inválido')
        return
    e[i] = e[i-1]
    e[i-1] = '__'
    return e

def der(e): 
    i = e.index('__')
    if(i in [3,7,11,15]):
        print('Movimiento inválido')
        return 0
    e[i] = e[i+1]
    e[i+1] = '__'
    return e

        
def sucesores(e):
    i = e.index('__')
    sucesores = []
        
    if(i not in [0,4,8,12]):
        iz = izq(e).copy()
        sucesores.append(iz)
        der(e)
        
    if(i not in [3,7,11,15]):
        de = der(e).copy()
        sucesores.append(de)
        izq(e)
            
    if(i not in [0,1,2,3]):
        up = arriba(e).copy()
        sucesores.append(up)
        abajo(e)
        
    if(i not in [12,13,14,15]):
        abaj = abajo(e).copy()
        sucesores.append(abaj)
        arriba(e)
    
    return sucesores
    
```

```python
def solucion(inicial, final):
    g = {}
    g[0] = 0
    abiertos = []
    cerrados = []
    abiertos.append(inicial)
    actual = abiertos[0]
    
    while(actual != final and len(abiertos) != 0):
        imprime(actual)
        abiertos.pop(0)
        cerrados.append(actual)
        
        hijos = sucesores(actual)
        
        #tratar repetidos:
        
        h_min = 17
        
        abiertos =  hijos
        actual = abiertos[0]
        for x in abiertos:
            if x in cerrados:
                hijos.remove(x)
                continue
            if h(x,final) <= h_min:
                print(h(x,final))
                h_min = h(x,final)
                actual = x
    cerrados.append(actual)
    
    print('Estado inicial: ')
    imprime(inicial)
    
    print('Estado final: ')
    imprime(final)
    
    print('Se resuelve en ', len(cerrados)-1, 'pasos:')
    i = 0
    for x in cerrados:
        if i == 0:
            i += 1
            continue
        print('\nPaso ',i)
        imprime(x)
        i += 1
        
```

```python
final = ['01','02','03','04','05','06','07','08','09','10','11','12','13','14','15','__']
inicial = ['01','02','03','04','05','06','07','__','09','10','12','08','13','14','11','15']
```

```python
wut = ['01','02','03','04','05','06','__','08','09','10','07','12','13','14','11','15']
print(len(wut))
imprime(wut)
scrs = sucesores(wut)
for x in scrs:
    if(x == final):
        continue
    print(h(x,final))
    imprime(x)
```
